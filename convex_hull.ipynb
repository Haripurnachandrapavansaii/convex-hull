```python
import math
import matplotlib.pyplot as plt

def orientation(p1, p2, p3):
    x1, y1, x2, y2, x3, y3 = *p1, *p2, *p3
    d = (y3-y2)*(x2-x1) - (y2-y1)*(x3-x2)
    if d > 0:
        return 1
    elif d < 0:
        return -1
    else:
        return 0
      

def dist(p1, p2):
    x1, y1, x2, y2 = *p1, *p2
    return math.sqrt((y2-y1)**2 + (x2-x1)**2)
  

def convex_hull (points):
    on_hull = min(points)
    hull = []
    while True:
        hull.append(on_hull)
        next_point = points[0]
        for point in points:
            o = orientation(on_hull, next_point, point)
            if next_point == on_hull or o == 1 or (o == 0 and dist(on_hull, point) > dist(on_hull, next_point)):
                next_point = point
        on_hull = next_point
        if on_hull == hull[0]:
            break
    return hull
points = [(1, 1), (2, 4), (3, 2), (4, 5), (5, 3),(0,0),(1,0),(2,3),(8,9),(-9,6)]
hull = convex_hull(points)
plt.scatter(*zip(*points))
lines = [(hull[i], hull[(i+1) % len(hull)]) for i in range(len(hull))]
for line in lines:
    plt.plot(*zip(*line), color='green')
plt.show()
```
